```python
#load the data

import pandas as pd
import numpy as np
from IPython.display import display, HTML

filename = 'targetedfast.xlsx'
sheetname = 'Sheet1'

xl = pd.ExcelFile(filename)
df1 = xl.parse(sheetname)

display(df1)

```

```python
currentlist = list(df1['Condition'])

order_wanted = ['10s; 10s', '10s; 20s', '10s; 30s', '10s; 60s']
std_ind = [0, 1]

nc_order = ['10s;','30s;']

zumit = []
zumit_nc = []

for item in order_wanted:
    zumit.append([index for index, value in enumerate(currentlist) if value == item])

for item in nc_order:
    zumit_nc.append([index for index, value in enumerate(currentlist) if value == item])

desired_xs = [0,1,2,3,6]
desired_xs_nc = [0,1,3]

raw_xs_nc=[1,1,3,3]
raw_xs=[1,1,2,2,3,3,6,6]
    
print(zumit)
print(zumit_nc)
```

```python
import matplotlib.pyplot as plt

ODs = np.array(df1['OD'])

compounds = ['DHAP', 'FBP', 'GAP']


#1.0 OD600 = 0.62 gCDW/l, 1.9 Ã— 10âˆ’11 gCDW per cell and 5.6 Ã— 10âˆ’14 l aqueous volume per cell

rabin = []
ourdata = []

for index, compound in enumerate(compounds):
    raw = np.array(df1[compound])/(ODs) * 1/(0.62*0.002)*1.9e-11/(5.6e-14)
    stdval = np.mean([raw[std_ind[0]], raw[std_ind[1]]])
    mean = [np.mean([raw[item[0]], 1/2*raw[item[1]]]) for item in zumit]
    mean_nc = [np.mean([raw[item[0]], 1/2*raw[item[1]]]) for item in zumit_nc]
    mean = np.log2(mean/stdval * 100*1e-6)
    mean_nc = np.log2(mean_nc/stdval * 100*1e-6)
    
    raw_plot=np.array([[raw[item[0]]/stdval * 100*1e-6, 1/2*raw[item[1]]/stdval * 100*1e-6] for item in zumit])    
    raw_plot_nc=np.array([[raw[item[0]]/stdval * 100*1e-6, 1/2*raw[item[1]]/stdval * 100*1e-6] for item in zumit_nc])
    
    zero_value = [np.mean(mean_nc)]
    
    
    mean = list(zero_value) + list(mean)
    mean_nc = list(zero_value) + list(mean_nc)


    stderr = [np.std([np.log2(raw[item[0]]/stdval * 100*1e-6), np.log2(1/2*raw[item[1]]/stdval * 100*1e-6)])/np.sqrt(2) for item in zumit]

    
    stderr = list([0]) + list(stderr)
    
    stderr_nc = [np.std([np.log2(raw[item[0]]/stdval * 100*1e-6), np.log2(1/2*raw[item[1]]/stdval * 100*1e-6)])/np.sqrt(2) for item in zumit_nc]
    
    stderr_nc = list([0]) + list(stderr_nc)
    
    #ax = plt.subplot(111)
    plt.figure()
    plt.errorbar(desired_xs, mean, yerr=stderr, label='SC')
    plt.plot(raw_xs, np.log2(raw_plot.flatten()),'o', label='SC')
    plt.errorbar(desired_xs_nc, mean_nc, yerr=stderr_nc, label='SCD')
    plt.plot(raw_xs_nc, np.log2(raw_plot_nc.flatten()),'o', label='SCD')
    plt.ylabel('log 2 change')
    #plt.xticks(desired_xs, order_wanted, rotation='vertical')
    plt.title(compound)
    plt.legend()
    
    #ax.xaxis.set_ticks(order_wanted)

    plt.show()
```
